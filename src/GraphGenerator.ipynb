```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
VERTEX_NUM = 11
```

```python
G = nx.erdos_renyi_graph(VERTEX_NUM, 0.5)
pos = nx.circular_layout(G)
plt.figure(200)
nx.draw(G, pos)
```

```python
G1 = nx.barabasi_albert_graph(VERTEX_NUM, 4)
pos = nx.circular_layout(G1)
plt.figure(201)
nx.draw(G1, pos)
```

```python
def generate_demand_matrix(graph: nx.Graph) -> np.array:
    nodes_num = len(graph.nodes)
    demand_matrix = np.zeros((nodes_num, nodes_num))
    for edge in G.edges:
        demand_matrix[edge[0], edge[1]] = np.random.randint(1, 10)
        demand_matrix[edge[1], edge[0]] = demand_matrix[edge[0], edge[1]]
    return demand_matrix
```

```python
generate_demand_matrix(G).tolist()
```

```python
generate_demand_matrix(G1).tolist()
```
